# Source Feature Count Sweep

Porownanie `price`, `Google`, `GDELT`, `Reddit` przy roznych liczbach cech na istniejacym `session_aligned` dataset.

Zasada:
- `price`: testujemy `top5`, `top10`, `top14`
- `Google`: `price14 + top1/top2/top3`
- `GDELT`: `price14 + top2/top4/top6`
- `Reddit`: `price14 + top2/top4/top7`

Kolejnosc cech w rodzinie jest zapisana jawnie w kodzie jako heurystyka, nie byla dobierana na teście.


```python
from source_feature_count_sweep_experiment import run_source_feature_count_sweep_experiment

result = run_source_feature_count_sweep_experiment()
result["best_by_source"].sort_values(["ticker", "model_name", "source_family"]).reset_index(drop=True)
```

```python
result["metrics"].query("split == 'test'").sort_values(["ticker", "model_name", "source_family", "n_features"]).reset_index(drop=True)
```

```python
result["metrics"].query("split == 'test'").pivot_table(
    index=["ticker", "model_name", "source_family"],
    columns="source_feature_count",
    values="balanced_accuracy",
)
```

```python
result["metrics"].query("split == 'test'").pivot_table(
    index=["ticker", "model_name", "source_family"],
    columns="source_feature_count",
    values="roc_auc",
)
```
