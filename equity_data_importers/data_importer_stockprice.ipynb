```python
from pathlib import Path
#!/usr/bin/env python3
import datetime as dt
import importlib
from equity_data_importers import config

importlib.reload(config)
from equity_data_importers.config import Config
import yfinance as yf

# 1) Make the end date inclusive for yf.download
yf_end = Config.END_DATE + dt.timedelta(days=1)

print(f"📉 Pobieram dane giełdowe: {Config.TICKER}")
stock_raw = yf.download(
    Config.TICKER,
    start=str(Config.START_DATE),
    end=str(yf_end),
    progress=False
)

# 2) Sanity check
if stock_raw.empty or "Close" not in stock_raw.columns or "Volume" not in stock_raw.columns:
    raise Exception("Błąd: brak danych lub brak kolumn 'Close' albo 'Volume'.")

# 3) Select & rename
stock_df = stock_raw[["Close", "Volume"]].copy()
stock_df.columns = ["stock_price", "stock_volume"]

print("\nKolumny stock_df:", stock_df.columns.tolist())
print("Zakres dat:", stock_df.index.min().date(), "→", stock_df.index.max().date())

# 4) Save
stock_df.to_csv("data/stock-prices-data.csv", index=True)
print("Dane zapisane do pliku: stock-prices-data.csv")
```
