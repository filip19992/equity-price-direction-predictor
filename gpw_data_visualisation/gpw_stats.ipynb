```python
import pandas as pd
import matplotlib.pyplot as plt

# Dane
df = pd.DataFrame({
    "instrument_group": ["Akcje/Equities"],
    "session_turnover_pln_thous": [470218203.46],
    "block_turnover_pln_thous": [21351561.23],
    "turnover_ratio_pct": [48.2]
})

# Przeliczenie: masz wartości w tys. PLN -> mld PLN
session_mld = df["session_turnover_pln_thous"][0] / 1e6
block_mld   = df["block_turnover_pln_thous"][0] / 1e6

# ===== WYKRES 1: Obrót sesyjny vs pakietowy (mld PLN) =====
plt.figure()
plt.bar(["Sesyjny", "Pakietowy"], [session_mld, block_mld])
plt.title("Struktura obrotu na rynku akcji")
plt.ylabel("Wartość obrotu (mld PLN)")
plt.xlabel("Rodzaj transakcji")
plt.xticks(rotation=0)

# wyłącza notację naukową na osi Y
plt.ticklabel_format(style='plain', axis='y')

plt.show()

# ===== WYKRES 2: Udział obrotu pakietowego =====
plt.figure()
plt.pie(
    [df["block_turnover_pln_thous"][0], df["session_turnover_pln_thous"][0]],
    labels=["Pakietowe", "Sesyjne"],
    autopct="%1.1f%%"
)
plt.title("Udział obrotu sesyjnego w rynku akcji")
plt.show()
```

```python
import matplotlib.pyplot as plt

session_mld = df["session_turnover_pln_thous"][0] / 1e6
block_mld   = df["block_turnover_pln_thous"][0] / 1e6

plt.figure()
bars = plt.bar(["Sesyjny", "Pakietowy"], [session_mld, block_mld])

plt.title("Struktura obrotu na rynku akcji")
plt.ylabel("Wartość obrotu (mld PLN)")
plt.xlabel("Rodzaj transakcji")
plt.ticklabel_format(style='plain', axis='y')

# wartości nad słupkami
for b in bars:
    y = b.get_height()
    plt.text(b.get_x() + b.get_width()/2, y, f"{y:.1f}", ha="center", va="bottom")

plt.show()
```
