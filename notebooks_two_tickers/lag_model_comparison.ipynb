# Two-Ticker Baseline: Model Family Comparison With MLP

This notebook compares model families on the same `TSLA + AAPL` panel.

Training modes:
- `pooled`
- `separate`

Feature variants:
- `base`
- `base_plus_all_lags`

Models:
- `logreg`
- `random_forest`
- `mlp_small`

Artifacts are written to `notebooks_two_tickers/outputs/lag_model_comparison`.


```python
import pandas as pd

from lag_model_comparison_experiment import run_lag_model_comparison_experiment

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 260)
```

```python
result = run_lag_model_comparison_experiment()

print(f"Dataset: {result['dataset_path']}")
print(f"Metrics: {result['metrics_path']}")
print(f"Predictions: {result['predictions_path']}")
print(f"Metadata: {result['metadata_path']}")
```

```python
test_all = result["metrics"].query("split == 'test' and ticker == 'ALL'").copy()
test_all = test_all.sort_values(["training_mode", "feature_variant", "model_name"]).reset_index(drop=True)
test_all
```

```python
summary = test_all.pivot_table(
    index=["training_mode", "feature_variant"],
    columns="model_name",
    values=["balanced_accuracy", "roc_auc"],
)
summary
```

```python
test_by_ticker = result["metrics"].query("split == 'test' and ticker != 'ALL'").copy()
test_by_ticker = test_by_ticker.sort_values(["ticker", "training_mode", "feature_variant", "model_name"]).reset_index(drop=True)
test_by_ticker
```
