# Stationary Alt Feature Experiment

Porownanie trzech wariantow cech:
- `price_only`
- `price_plus_alt_current`
- `price_plus_alt_stationary`

Nowy wariant `stationary` zostawia price side bez zmian, ale zamienia czesc alt-data z poziomow na:
- zmiany,
- rolling gaps,
- rolling z-scores,
- surprise features,
- source-separated Reddit sentiment.


```python
from stationary_alt_feature_experiment import run_stationary_alt_feature_experiment

result = run_stationary_alt_feature_experiment()
result["metrics"].query("split == 'test'").sort_values(["ticker", "model_name", "feature_set"]).reset_index(drop=True)
```
