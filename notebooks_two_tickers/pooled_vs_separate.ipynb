# Two-Ticker Baseline: Pooled vs Separate

This notebook compares two training modes on the same `TSLA + AAPL` panel:
- `pooled`: one model trained on both tickers together
- `separate`: one model trained per ticker

Feature sets:
- `price_only`
- `alt_only`
- `price_plus_alt`

Models:
- `logreg`
- `random_forest`

Artifacts are written to `notebooks_two_tickers/outputs/pooled_vs_separate`.


```python
import pandas as pd

from pooled_vs_separate_experiment import run_pooled_vs_separate_experiment

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 240)
```

```python
result = run_pooled_vs_separate_experiment()

print(f"Dataset: {result['dataset_path']}")
print(f"Metrics: {result['metrics_path']}")
print(f"Predictions: {result['predictions_path']}")
print(f"Feature sets: {result['feature_sets_path']}")
print(f"Metadata: {result['metadata_path']}")
```

```python
test_all = result["metrics"].query("split == 'test' and ticker == 'ALL'").copy()
test_all = test_all.sort_values(["training_mode", "model_name", "feature_set"]).reset_index(drop=True)
test_all
```

```python
test_per_ticker = result["metrics"].query("split == 'test' and ticker != 'ALL'").copy()
test_per_ticker = test_per_ticker.sort_values(["ticker", "training_mode", "model_name", "feature_set"]).reset_index(drop=True)
test_per_ticker
```
