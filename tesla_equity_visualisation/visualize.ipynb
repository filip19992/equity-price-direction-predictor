```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane
df = pd.read_csv('../equity_data_importers/data/stock-prices-data.csv')


# Konwertuj kolumnę Date na datetime
df['Date'] = pd.to_datetime(df['Date'])

# Utwórz wykres
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Wykres ceny akcji
ax1.plot(df['Date'], df['stock_price'], linewidth=2, color='#1f77b4')
ax1.set_title('Cena akcji Tesla', fontsize=14, fontweight='bold')
ax1.set_xlabel('Data')
ax1.set_ylabel('Cena ($)')
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# Wykres wolumenu obrotu
ax2.bar(df['Date'], df['stock_volume'], color='#ff7f0e', alpha=0.7)
ax2.set_title('Wolumin obrotu akcji Tesla', fontsize=14, fontweight='bold')
ax2.set_xlabel('Data')
ax2.set_ylabel('Wolumin')
ax2.grid(True, alpha=0.3, axis='y')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
